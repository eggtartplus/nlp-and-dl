````python
import requests
import torchaudio
import matplotlib.pyplot as plt
import torchaudio.functional as F
import torchaudio.transforms as T
import librosa

target_sample_rate = 16000
n_fft = 512

# a function to plot spectrogram
def plot_spectrogram(specgram, title=None, ylabel="freq_bin"):
    fig, axs = plt.subplots(1, 1)
    axs.set_title(title or "Spectrogram (db)")
    axs.set_ylabel(ylabel)
    axs.set_xlabel("frame")
    im = axs.imshow(librosa.power_to_db(specgram), origin="lower", aspect="auto")
    fig.colorbar(im, ax=axs)
    plt.show(block=False)

# url to find the audio sample
url = "https://drwuz.com/CSC3160/materials/audio/LJ001-0048.wav"
# request the url
r = requests.get(url)

# filename, where to save the audio
filename = "LJ001-0048.wav"
with open(filename, 'wb') as f:
    f.write(r.content)

# load audio sample to memory
waveform, sample_rate = torchaudio.load(filename)

# create an object that resamples a waveform
resampler = T.Resample(sample_rate, target_sample_rate, dtype=waveform.dtype)
# resample a waveform from its original sampling rate to the target sampling rate
resampled_waveform = resampler(waveform) 

# create an object that extract spectrogram
spec_trans = T.Spectrogram(n_fft=n_fft, power=2)

# extract spectrogram
spec = spec_trans(resampled_waveform)

print("Shape of waveform: {}".format(waveform.size()))
print("Sample rate of waveform: {}".format(sample_rate))

print("Shape of resampled_waveform: {}".format(resampled_waveform.size()))

plt.figure()
plt.plot(waveform.t().numpy())

plot_spectrogram(spec[0])
````

In this part, you need to extract power spectorgram (FFT 1024) and 80-dimensional mel-spectrogram (80 Mel Filter Banks) with 5ms, 10ms, 15ms frameshift, with the same window size 30ms. The speech signal is at 16 kHz sampling rate.
* Visualize each of the power spectorgram (FFT 1024) and 80-dimensional mel-spectrogram
* Plot the spectrum of the 150th frame of the 512-dimensional spectrogram with 10 ms frameshift



````python
# Write your implementation here
import numpy as np
# 1.power spectorgram (FFT 1024) with 5ms frameshift
spec_trans1 = T.Spectrogram(n_fft=1024, power=2, win_length=int(target_sample_rate*0.03), hop_length=int(target_sample_rate*0.005))
spec1 = spec_trans1(resampled_waveform)
plot_spectrogram(spec1[0])

# 2.power spectorgram (FFT 1024) with 10ms frameshift
spec_trans2 = T.Spectrogram(n_fft=1024, power=2, win_length=int(target_sample_rate*0.03), hop_length=int(target_sample_rate*0.01))
spec2 = spec_trans2(resampled_waveform)
plot_spectrogram(spec2[0])

# 3.power spectorgram (FFT 1024) with 15ms frameshift
spec_trans3 = T.Spectrogram(n_fft=1024, power=2, win_length=int(target_sample_rate*0.03), hop_length=int(target_sample_rate*0.015))
spec3 = spec_trans3(resampled_waveform)
plot_spectrogram(spec3[0])

# 4.80-dimensional mel-spectrogram (80 Mel Filter Banks) with 5ms frameshift
meltransform1 = T.MelSpectrogram(target_sample_rate, n_mels=80, win_length=int(target_sample_rate*0.03), hop_length=int(target_sample_rate*0.005), n_fft=1024)
mel_specgram1 = meltransform1(waveform) 
plot_spectrogram(mel_specgram1[0])

# 5.80-dimensional mel-spectrogram (80 Mel Filter Banks) with 10ms frameshift
meltransform2 = T.MelSpectrogram(target_sample_rate, n_mels=80, win_length=int(target_sample_rate*0.03), hop_length=int(target_sample_rate*0.01), n_fft=1024)
mel_specgram2 = meltransform2(waveform) 
plot_spectrogram(mel_specgram2[0])

# 6.80-dimensional mel-spectrogram (80 Mel Filter Banks) with 15ms frameshift
meltransform3 = T.MelSpectrogram(target_sample_rate, n_mels=80, win_length=int(target_sample_rate*0.03), hop_length=int(target_sample_rate*0.015), n_fft=1024)
mel_specgram3 = meltransform3(waveform) 
plot_spectrogram(mel_specgram3[0])

# 7.the spectrum of the 150th frame of the 512-dimensional spectrogram with 10 ms frameshift
hop = int(target_sample_rate*0.01)
amplitude = resampled_waveform.t().numpy()[hop*149:hop*149+int(target_sample_rate*0.03)]
# Frequency domain representation
fourierTransform = 2*np.fft.fft(amplitude)/len(amplitude)           # Normalize amplitude
fourierTransform = fourierTransform[range(int(len(amplitude)/2))]       # Exclude sampling frequency

tpCount     = len(amplitude)
values      = np.arange(int(tpCount/2))
timePeriod  = tpCount/target_sample_rate
frequencies = values/timePeriod

plt.figure(figsize=(12, 5))
plt.plot(frequencies, abs(fourierTransform), linewidth=2)
plt.xlabel('Frequency', fontsize=18)
plt.ylabel('Amplitude', fontsize=18)
plt.title('Fourier transform depicting the frequency components', fontsize=18)
````
